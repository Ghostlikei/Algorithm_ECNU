```python
import csv
import numpy as np
import matplotlib.pyplot as plt

x_axis = []
y1_axis = []
y2_axis = []
plt.figure(figsize=(10,8),dpi=200)

with open("timeData1.csv", "r") as csvfile:
    reader = list(csv.reader(csvfile))
    for row in reader[1:]:
        x_axis.append(int(row[0]))
        y1_axis.append(float(row[1]))
        
with open("timeData.csv", "r") as csvfile:
    reader = list(csv.reader(csvfile))
    for row in reader[1:]:
        y2_axis.append(float(row[1]))

l1, = plt.plot(x_axis,y1_axis,label = "QuickSort")


plt.xlabel("size")
plt.ylabel("time")

l2, = plt.plot(x_axis,y2_axis,label = "RandomizedQS")


plt.legend(handles = [l1,l2], labels = ["QuickSort","RandomizedQS"])
plt.show()
```
