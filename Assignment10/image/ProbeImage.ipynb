```python
import csv
import numpy as np
import matplotlib.pyplot as plt

x_axis = []
y1_axis = []
y2_axis = []
y3_axis = []
plt.figure(figsize=(10,8),dpi=200)

with open("time.csv", "r") as csvfile:
    reader = list(csv.reader(csvfile))
    for row in reader[1:]:
        x_axis.append(int(row[0]))
        y1_axis.append(float(row[1]))
        y2_axis.append(float(row[2]))
        y3_axis.append(float(row[3]))




plt.xlabel("queryTimes")
plt.ylabel("time(ms)")
l1, = plt.plot(x_axis,y1_axis)
l2, = plt.plot(x_axis,y2_axis)
l3,= plt.plot(x_axis,y3_axis)


plt.legend(handles = [l1,l2,l3], labels = ["OptimalBST","Array","Hashtable"])
plt.show()
```
