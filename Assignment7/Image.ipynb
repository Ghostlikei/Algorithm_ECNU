```python
import csv
import numpy as np
import matplotlib.pyplot as plt

x_axis = []
y1_axis = []
y2_axis = []
plt.figure(figsize=(10,8),dpi=200)

with open("Data.csv", "r") as csvfile:
    reader = list(csv.reader(csvfile))
    for row in reader[1:]:
        x_axis.append(int(row[0]))
        y1_axis.append(float(row[1]))
        y2_axis.append(float(row[2]))



plt.xlabel("size")
plt.ylabel("time")
l1 = plt.scatter(x_axis,y1_axis,s=15)
l2 = plt.scatter(x_axis,y2_axis,s=15)


plt.legend(handles = [l1,l2], labels = ["Insert","Find(100 entries)"])
plt.show()
```
