```python
import csv
import numpy as np
import matplotlib.pyplot as plt

x_axis = []
y1_axis = []
y2_axis = []
plt.figure(figsize=(10,8),dpi=1500)

with open("data.csv", "r") as csvfile:
    reader = list(csv.reader(csvfile))
    for row in reader[1:]:
        x_axis.append(int(row[0]))
        y1_axis.append(float(row[1]))
        
with open("set_heap.csv", "r") as csvfile:
    reader = list(csv.reader(csvfile))
    for row in reader[1:]:
        y2_axis.append(float(row[1]))

# y1 = np.array(y1_axis)
# y2 = np.array(y2_axis)
# x = np.array(x_axis)

# f1 = np.polyfit(x, y1, 1)
# p1 = np.poly1d(f1)
# yvals1 = p1(x)

# f2 = np.polyfit(x, y2, 1)
# p2 = np.poly1d(f2)
# yvals2 = p2(x)


l1, = plt.plot(x_axis,y1_axis,label = "heapSort")


plt.xlabel("size")
plt.ylabel("time")

l2, = plt.plot(x_axis,y2_axis,label = "setHeap")


plt.legend(handles = [l1,l2], labels = ["heapSort","setHeap"])
plt.show()
```
